```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
#request url
url = "https://en.wikipedia.org/wiki/COVID-19_lockdowns"
wiki_class = "wikitable sortable"
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text,"html.parser")
locktable = soup.find('table',class_="wikitable")
# locktable
```

```python
df=pd.read_html(str(locktable))
# df
```

```python
df[0].iloc[91,:]
```

```python
df_nat = pd.DataFrame(df[0].iloc[90,:])
df_nat
```

```python
df_par = pd.DataFrame(df[0].iloc[91,:])
df_par
# print(df_par.columns)
```

```python
df_par_clean = df_par.reset_index().drop('level_0',axis=1)
df_par_clean
```

```python
df_par_clean1 = df_par_clean.rename(columns={91:"values"})
df_par_clean1
```

```python
df_par_clean2 = pd.pivot(df_par_clean1,columns=["level_1","level_2"],values="values")
df_par_clean2
```

```python
# df_par_clean2.rename_axis(None,axis=1,names=['level_1','level_2'])
# df_par_clean2.index
# df_par_clean2.head()
```

```python
df_par_clean3 = df_par_clean2.copy()
df_par_clean3.iloc[0,:] = df_par_clean1['values']
df_par_clean3.head()
```

```python
df_par_clean3 = df_par_clean3.drop(labels=range(1,len(df_par_clean3)),axis=0)
df_par_clean3
```

```python
df_nat_clean = df_nat.reset_index().drop('level_0',axis=1)
df_nat_clean1 = df_nat_clean.rename(columns={90:"values"})
df_nat_clean2 = pd.pivot(df_nat_clean1,columns=["level_1","level_2"],values="values")
df_nat_clean2.head(5)
```

```python
df_nat_clean3 = df_nat_clean2.copy()
df_nat_clean3.iloc[0,:] = df_nat_clean1['values']
df_nat_clean3.head()
```

```python
df_nat_clean3 = df_nat_clean3.drop(labels=range(1,len(df_nat_clean3)),axis=0)
df_nat_clean3
```

```python
df_nat_clean3.set_index(['Country / territory'])
```

```python
df_nat_clean3 = df_nat_clean3.replace(to_replace ='\[.*', value = '', regex = True)
df_nat_clean3
```

```python
df_par_clean3 = df_par_clean3.replace(to_replace ='\[.*', value = '', regex = True)
df_par_clean3
```
