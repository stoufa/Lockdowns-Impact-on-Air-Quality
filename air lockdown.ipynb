## Uganda Covid-19 Air quality and lockdowns

```python
df_ug = pd.read_csv("Ugandacovid.csv",low_memory = False,parse_dates=True)
df_ug.head(5)
```

```python
df_ug['Specie'].value_counts()
```

```python
df_ug.info()
```

```python
for col in ['Date','start date','end date']:
    df_ug[col] = pd.to_datetime(df_ug[col])
df_ug.head()
```

```python
df_ug.tail()
```

```python
df_ug[df_ug['Specie']=='temperature'].loc[:30,'median']
```

```python
df_ug[df_ug['Specie']=='temperature'].loc[:30,'Date']
```

```python
plt.figure(figsize=[12,8])
plt.plot(df_ug[df_ug['Specie']=='temperature'].loc[:30,'Date'], df_ug[df_ug['Specie']=='temperature'].loc[:30,'median'])
plt.show()
```
